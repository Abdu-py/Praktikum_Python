## 3. Program
### 3.1 List

```

fruits = ['apple', 'banana', 'cherry']
fruits.append('orange')
print(fruits[0])
fruits[1] = 'blueberry'
fruits.remove('cherry')
print(fruits)
```

### 3.2 Tuple

```

coordinates = (10, 20, 30)
print(coordinates[0])
```

### 3.3 Set

```

fruits_set = {'apple', 'banana', 'cherry'}
fruits_set.add('orange')
fruits_set.add('apple')
print(fruits_set)
```

### 3.4 Dictionary

```

student = {'name': 'John Doe', 'age': 22, 'major': 'Computer Science'}
print(student['name'])
student['graduation_year'] = 2023
del student['age']
print(student)
```

### 3.5 Data Training AI

```

X = [[1,2],[3,4],[5,6]]
y = [0,1,0]
print("Fitur:", X)
print("Label:", y)
```

### 3.6 Parameter Model AI

```

model_params = {'weights':[0.5,-0.2],'bias':0.1}
print(model_params['weights'])
print(model_params['bias'])
model_params['bias']=0.2
print(model_params)
```
