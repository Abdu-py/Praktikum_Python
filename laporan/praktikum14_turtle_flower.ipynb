Praktikum ini bertujuan untuk mempelajari penggunaan modul **turtle** pada Python dalam membuat visualisasi grafis berbentuk pola bunga.

## Dasar Teori
Turtle Graphics adalah modul standar Python yang digunakan untuk pembelajaran grafika dasar dan pemrograman visual.

## Program 1: Flower 1

```python
import turtle as tur
import colorsys as cs
tur.setup(600,600)
tur.speed(0)
tur.tracer(100)
tur.width(2)
tur.screensize(70)
tur.bgcolor('black')
for j in range (25):
 for i in range (15) :
    tur.color(cs.hsv_to_rgb(i/15,j/25,1))
    tur.right(90)
    tur.circle(200-j*4,90)
    tur.left(90)
    tur.circle(200-j*4,90)
    tur.right(180)
    tur.circle(2,54)
    tur.hideturtle()
    tur.done
```

## Program 2: Flower 2

```python
import turtle
t = turtle.Turtle()
for i in range (180):
    t.speed(0)
    t.color('#277BC0')
    t.circle(190 - i, 90)
    t.left(90)
    t.color('#E38B29')
    t.circle(190 - i, 90)
    t.left(18)
```

## Program 3: Flower 3

```python
import turtle as tr

s = tr.Screen()

tr.setup(800, 800)
s.bgcolor('#262626')
tr.pencolor('#540d6e')
tr.speed(0)
tr.tracer(100)
tr.pensize(1)
col = ('#FF7F3F', '#FBDF07', '#89CFFD', '#F94892')

for i in range(3):
    for n in range(400):
        tr.color(col[n%4])
        tr.circle(190-n/2,90)
        tr.left(90)
        tr.circle(190-n/2,90)
        tr.color(col[n%4])
    tr.left(30)
s.exitonclick()
```

## Program 4: Flower 4

```python
import turtle as tur
import colorsys as cs

tur.setup(800,800)
tur.speed(0)
tur.width(2)
tur.bgcolor("black")
for j in range(25):
    for i in range(15):
        tur.color(cs.hsv_to_rgb(i/15,j/25,1))
        tur.right(90)
        tur.circle(200-j*4,90)
        tur.left(90)
        tur.circle(200-j*4,90)
        tur.right(180)
        tur.circle(50,24)
tur.hideturtle()
tur.done()
```

## Program 5: Flower 5

```python
import turtle

t = turtle.Turtle()
s = turtle.Screen()
s.bgcolor('black')
t.speed(0)

for i in range(17):
    t.color('purple')
    t.begin_fill()
    t.circle(190-i, 90)
    t.left(98)
    t.end_fill()
    t.color('pink')
    t.begin_fill()
    t.circle(190-i, 90)
    t.left(18)
    t.end_fill()
s.exitonclick()
```
